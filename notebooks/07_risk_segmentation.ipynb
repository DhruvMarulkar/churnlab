```python
import pandas as pd

risk_df = pd.read_csv("../data/processed/churn_risk_raw.csv")

X_train = pd.read_csv("../data/processed/X_train.csv")

print(risk_df.head())
print(X_train.head())
```

```python
def categorize_risk(p):
    if p > 0.7:
        return "High"
    elif p > 0.4:
        return "Medium"
    else:
        return "Low"
risk_df["risk_segment"] = risk_df["churn_probability"].apply(categorize_risk)
risk_df.head()
```

```python
seg_df = pd.concat([X_train, risk_df], axis=1)
seg_df.head()
```

```python
seg_df["risk_segment"].value_counts()
```

```python
seg_df["risk_segment"].value_counts(normalize=True)*100
```

```python
seg_df.to_csv("../data/processed/churn_risk_segmented.csv", index=False)
print("Saved segmentation file to data/processed/churn_risk_segmented.csv")
```

## 📌 Risk Segmentation Summary

We converted raw churn probabilities into business-friendly risk tiers:

- High Risk (> 0.7)
- Medium Risk (0.4–0.7)
- Low Risk (< 0.4)

This segmentation layer enables targeted retention strategies and prepares the dataset for the Power BI dashboard and API deployment.
